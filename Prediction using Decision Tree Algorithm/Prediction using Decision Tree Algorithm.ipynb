**Task #6 : Prediction using Decision Tree Algorithm**

Importing libraries

```python
import sklearn.datasets as datasets
import pandas as pd
```

Importing dataset

```python
iris=datasets.load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
X.head()
```

```python
X.tail()
```

```python
X.info()
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
Y = iris.target
Y
```


Split dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Defining the Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)

print('Decision Tree Classifer Created Successfully')
```

```python
y_predict = classifier.predict(X_test)
```

Constructing confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

```python
classifier.predict([[4.9, 3.0, 4.1, 0.3], [4.1, 0.2, 3.0, 0.1]])
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

Visualizing the Decision tree

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(classifier, feature_names = fn, class_names = cn, filled = True);
```
